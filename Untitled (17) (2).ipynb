```python
class FCalculator:
    def __init__(self, a, b, eta, rho, D, h, ks, m_u, N_T=10):
        self.a = a
        self.b = b
        self.eta = eta
        self.rho = rho
        self.D = D
        self.h = h
        self.m_u = m_u
        self.nu = np.sqrt(ks / m_u)
        self.nt = N_T

    def S_f(self, k_b, kx_b, ky_a):
        s_e = np.sin(self.eta)
        ab_s = (self.a / self.b) * s_e
        
        out = 0j
        dl = np.cos(s_e * ky_a)

        for n in range(-self.nt, self.nt + 1):
            kxn = kx_b + 2.0 * np.pi * n
            
            lam = np.sqrt((kxn / k_b)**2 - 1.0 + 0j)
            gam = np.sqrt((kxn / k_b)**2 + 1.0 + 0j)
            
            t1 = np.sinh(gam * k_b * ab_s) / (dl - np.cosh(gam * k_b * ab_s)) / gam
            t2 = np.sinh(lam * k_b * ab_s) / (dl - np.cosh(lam * k_b * ab_s)) / lam
            
            out += t1 - t2
            
        return out

    def F_meta(self, f_hz, kx, ky):
        w = 2.0 * np.pi * f_hz
        kap = (self.rho * self.h * w**2 / self.D) ** 0.25
        
        dn = 1.0 - (w / self.nu)**2
        if abs(dn) < 1e-10:
            return np.nan
            
        mu = self.m_u * w**2 / (self.D * dn)
        
        lhs = 4.0 * self.b * kap**3 / mu
        rhs = self.S_f(kap * self.b, kx * self.b, ky * self.a)
        
        return np.real(lhs - rhs)
```

```python
class TorchPhysicsResidual(nn.Module):
    def __init__(
        self, a, b, eta, rho, D, h, ks, m_u, 
        N_T=10, eps=1e-7, sqrt_damping=1e-5, 
        denom_damping=1e-3, resonance_exclusion_hz=15.0
    ):
        super().__init__()

        self.a = float(a)
        self.b = float(b)
        self.eta = float(eta)
        self.rho = float(rho)
        self.D = float(D)
        self.h = float(h)
        self.ks = float(ks)
        self.m_u = float(m_u)
        self.nu = float(np.sqrt(ks / m_u))
        self.f_r = self.nu / (2.0 * np.pi)
        self.nt = int(N_T)
        self.eps = float(eps)

        self.sq_dmp = float(sqrt_damping)
        self.dn_dmp = float(denom_damping)
        self.res_exc = float(resonance_exclusion_hz)

    def S_f(self, k_b, kx_b, ky_a):
        c_dt = torch.complex64

        k_bc = k_b.to(c_dt)
        kx_bc = kx_b.to(c_dt)
        ky_ac = ky_a.to(c_dt)

        s_eta_r = np.sin(self.eta)

        s_eta = torch.tensor(s_eta_r, dtype=torch.float32, device=k_b.device).to(c_dt)
        ab_s_eta = torch.tensor((self.a / self.b) * s_eta_r, dtype=torch.float32, device=k_b.device).to(c_dt)

        dn_L = torch.cos(s_eta * ky_ac)
        out = torch.zeros_like(k_bc)

        sq_sh = torch.tensor(1j * self.sq_dmp, dtype=c_dt, device=k_b.device)
        dn_sh = torch.tensor(1j * self.dn_dmp, dtype=c_dt, device=k_b.device)

        for n in range(-self.nt, self.nt + 1):
            kxn = kx_bc + 2.0 * np.pi * n
            rt = kxn / (k_bc + self.eps)

            l_arg = rt**2 - 1.0
            g_arg = rt**2 + 1.0

            lam_n = torch.sqrt(l_arg + sq_sh)
            gam_n = torch.sqrt(g_arg + sq_sh)

            zg = gam_n * k_bc * ab_s_eta
            zl = lam_n * k_bc * ab_s_eta

            dg = dn_L - torch.cosh(zg) + dn_sh
            dl = dn_L - torch.cosh(zl) + dn_sh

            t1 = torch.sinh(zg) / dg / (gam_n + self.eps)
            t2 = torch.sinh(zl) / dl / (lam_n + self.eps)

            out = out + t1 - t2

        return out

    def res_diag(self, f_hz, kx, ky):
        w = 2.0 * np.pi * f_hz
        kap = (self.rho * self.h * w**2 / self.D).clamp_min(self.eps).pow(0.25)
        
        dn = 1.0 - w**2 / self.nu**2
        dn_sf = torch.where(torch.abs(dn) < 1e-4, torch.sign(dn + self.eps) * 1e-4, dn)

        mu = self.m_u * w**2 / (self.D * dn_sf)

        lhs = 4.0 * self.b * kap**3 / (mu + self.eps)
        rhs = self.S_f(kap * self.b, kx * self.b, ky * self.a)
        rhs_r = torch.real(rhs)

        raw_r = lhs - rhs_r
        
        scl = torch.abs(lhs.detach()) + torch.abs(rhs_r.detach()) + 1.0
        r_norm = raw_r / scl 

        return {
            "f": f_hz,
            "r": r_norm,
            "l": lhs,
            "r_r": rhs_r
        }

    def residual_norm(self, f_hz, kx, ky):
        return self.res_diag(f_hz, kx, ky)["r"]

    def physics_loss(self, f_hz, kx, ky):
        r = self.res_diag(f_hz, kx, ky)["r"]
        msk = torch.abs(f_hz - self.f_r) > self.res_exc

        if msk.any():
            return torch.mean(r[msk]**2)

        return torch.mean(r * 0.0)

    @torch.no_grad()
    def sing_diag(self, f_hz, kx, ky):
        w = 2.0 * np.pi * f_hz
        kap = (self.rho * self.h * w**2 / self.D).clamp_min(self.eps).pow(0.25)

        k_b = kap * self.b
        kx_b = kx * self.b
        ky_a = ky * self.a

        m_l_arg = float("inf")
        m_dl = float("inf")
        m_dg = float("inf")
        w_n = None

        s_eta_r = np.sin(self.eta)
        ab_s_eta = (self.a / self.b) * s_eta_r

        dn_L = torch.cos(torch.tensor(s_eta_r, dtype=torch.float32, device=kx.device) * ky_a)

        for n in range(-self.nt, self.nt + 1):
            kxn = kx_b + 2.0 * np.pi * n
            rt = kxn / (k_b + self.eps)

            l_arg = rt**2 - 1.0
            g_arg = rt**2 + 1.0

            lam_n = torch.sqrt(l_arg.to(torch.complex64) + 1j * self.sq_dmp)
            gam_n = torch.sqrt(g_arg.to(torch.complex64) + 1j * self.sq_dmp)

            zl = lam_n * k_b.to(torch.complex64) * ab_s_eta
            zg = gam_n * k_b.to(torch.complex64) * ab_s_eta

            dl = dn_L.to(torch.complex64) - torch.cosh(zl)
            dg = dn_L.to(torch.complex64) - torch.cosh(zg)

            c_l = torch.min(torch.abs(l_arg)).item()
            c_dl = torch.min(torch.abs(dl)).item()
            c_dg = torch.min(torch.abs(dg)).item()

            if c_l < m_l_arg:
                m_l_arg = c_l
                w_n = n

            m_dl = min(m_dl, c_dl)
            m_dg = min(m_dg, c_dg)

        print("sing_diag res:")
        print(f" min |l_arg|: {m_l_arg:.4e} (n={w_n})")
        print(f" min |dl|: {m_dl:.4e}")
        print(f" min |dg|: {m_dg:.4e}")
```

```python
def t_stats(nm, x):
    xd = x.detach()
    fin = torch.isfinite(xd)

    print(f"{nm} | shp:{tuple(xd.shape)} | fin:{fin.sum().item()}/{xd.numel()}")

    if fin.any():
        xf = xd[fin]
        print(f"  min:{xf.min().item():.2e} max:{xf.max().item():.2e} avg:{xf.mean().item():.2e}")
        if xf.numel() > 1:
            print(f"  std:{xf.std().item():.2e}")

    if (~fin).any():
        bad = torch.nonzero(~fin, as_tuple=False)
        print(f"  ERR:{bad.shape[0]} | idx:{bad[:5].detach().cpu().numpy()}")


def dump_grads(mdl):
    print("--- grad info ---")
    tot_sq = 0.0
    mx_g = 0.0
    has_g = False

    for nm, p in mdl.named_parameters():
        if p.grad is None:
            continue

        g = p.grad.detach()
        fin = torch.isfinite(g)

        if not fin.all():
            print(f"  NAN in: {nm}")
            t_stats(f"g[{nm}]", g)

        if fin.any():
            gf = g[fin]
            p_nrm = torch.norm(gf).item()
            tot_sq += p_nrm**2
            mx_g = max(mx_g, torch.max(torch.abs(gf)).item())
            has_g = True

    if has_g:
        t_nrm = tot_sq ** 0.5
        print(f"  tot_nrm: {t_nrm:.2e} | max_g: {mx_g:.2e}")
    else:
        print("  no grads")


def dbg_fail(mdl, ep, b_id, kb, fq, phys, l_p, l_o, l_t):
    print(f"\n!!! FAIL | ep:{ep} b:{b_id} !!!")

    kx = kb[:, 0:1]
    ky = kb[:, 1:2]

    v_p = l_p.detach().cpu().item() if torch.numel(l_p) == 1 else l_p
    v_o = l_o.detach().cpu().item() if torch.numel(l_o) == 1 else l_o
    v_t = l_t.detach().cpu().item() if torch.numel(l_t) == 1 else l_t

    print(f"L -> p:{v_p} o:{v_o} t:{v_t}")

    t_stats("fq", fq)

    with torch.no_grad():
        for j in range(fq.shape[1]):
            fj = fq[:, j]
            print(f"  br{j} -> [{fj.min().item():.1f}, {fj.max().item():.1f}] a:{fj.mean().item():.1f}")

        df = fq[:, 1:] - fq[:, :-1]
        t_stats("df", df)

        for j in range(fq.shape[1]):
            print(f"\n-- br{j} diag --")
            fj = fq[:, j:j+1]
            
            # Using updated names from the refactored TorchPhysicsResidual
            dg = phys.res_diag(fj, kx, ky) 

            t_stats("r", dg["r"])
            t_stats("l", dg["l"])
            t_stats("r_r", dg["r_r"])
            phys.sing_diag(fj, kx, ky)
            
    print("!!! END FAIL !!!\n")
```

```python
def gen_k_path(n_sg=80, a=1.0):
    km = np.pi / a

    x_gx = np.linspace(0, km, n_sg, endpoint=False)
    y_gx = np.zeros(n_sg)

    x_xm = np.full(n_sg, km)
    y_xm = np.linspace(0, km, n_sg, endpoint=False)

    x_mg = np.linspace(km, 0, n_sg, endpoint=True)
    y_mg = np.linspace(km, 0, n_sg, endpoint=True)

    rt_x = np.concatenate([x_gx, x_xm, x_mg])
    rt_y = np.concatenate([y_gx, y_xm, y_mg])

    return rt_x, rt_y

class MixedKData(Dataset):
    def __init__(self, n_rnd, rt_x, rt_y, a=1.0, rt_os=100, sd=42, rt_j=1e-4):
        rng = np.random.default_rng(sd)
        km = np.pi / a

        x_rnd = rng.uniform(0.0, km, size=n_rnd)
        y_rnd = rng.uniform(0.0, km, size=n_rnd)

        x_rep = np.tile(rt_x, rt_os)
        y_rep = np.tile(rt_y, rt_os)

        x_rep += rng.normal(0.0, rt_j, size=x_rep.shape)
        y_rep += rng.normal(0.0, rt_j, size=y_rep.shape)

        x_rep = np.clip(x_rep, 0.0, km)
        y_rep = np.clip(y_rep, 0.0, km)

        x = np.concatenate([x_rnd, x_rep])
        y = np.concatenate([y_rnd, y_rep])

        self.k = torch.tensor(np.column_stack([x, y]), dtype=torch.float32)

    def __len__(self):
        return len(self.k)

    def __getitem__(self, idx):
        return self.k[idx]

n_sg = 80
rt_x, rt_y = gen_k_path(n_sg=n_sg, a=a)

ds_trn = MixedKData(
    n_rnd=n_trn,
    rt_x=rt_x,
    rt_y=rt_y,
    a=a,
    rt_os=rt_os,
    sd=sd,
    rt_j=2e-4
)

ds_val = MixedKData(
    n_rnd=n_val,
    rt_x=rt_x,
    rt_y=rt_y,
    a=a,
    rt_os=max(20, rt_os // 4),
    sd=sd + 1,
    rt_j=2e-4
)

is_gpu = (dv.type == "cuda")

dl_trn = DataLoader(
    ds_trn,
    batch_size=bsz,
    shuffle=True,
    num_workers=0,
    pin_memory=is_gpu
)

dl_val = DataLoader(
    ds_val,
    batch_size=bsz,
    shuffle=False,
    num_workers=0,
    pin_memory=is_gpu
)

print(f"rt_pts: {len(rt_x)}")
print(f"trn_pts: {len(ds_trn)}")
print(f"val_pts: {len(ds_val)}")
```

```python
def inv_sig(y, e=1e-6):
    y = np.clip(y, e, 1.0 - e)
    return np.log(y / (1.0 - y))

class FEmb(nn.Module):
    def __init__(self, in_d=2, emb_d=64, scl=1.5):
        super().__init__()
        self.B = nn.Parameter(torch.randn(in_d, emb_d) * scl, requires_grad=False)

    def forward(self, x):
        xp = 2.0 * np.pi * (x @ self.B)
        return torch.cat([torch.sin(xp), torch.cos(xp)], dim=-1)

class P5BPinn(nn.Module):
    def __init__(self, kmx, f0=1.0, f1=800.0, brs=5, hdn=256):
        super().__init__()
        self.kmx = float(kmx)
        self.f0 = float(f0)
        self.f1 = float(f1)
        self.brs = int(brs)

        self.emb = FEmb(in_d=2, emb_d=64, scl=1.5)

        self.b_lst = nn.ModuleList([
            nn.Sequential(
                nn.Linear(128, hdn),
                nn.Mish(),
                nn.Linear(hdn, hdn),
                nn.Mish(),
                nn.Linear(hdn, hdn),
                nn.Mish(),
                nn.Linear(hdn, 1)
            ) for _ in range(self.brs)
        ])

        self.init_hds()

    def init_hds(self):
        tf = np.linspace(
            self.f0 + 0.1 * (self.f1 - self.f0),
            self.f1 - 0.1 * (self.f1 - self.f0),
            self.brs
        )
        y = (tf - self.f0) / (self.f1 - self.f0)
        bs = inv_sig(y)

        for j, b in enumerate(self.b_lst):
            ll = b[-1]
            nn.init.zeros_(ll.weight)
            with torch.no_grad():
                ll.bias.copy_(torch.tensor([bs[j]], dtype=torch.float32))

    def forward(self, k):
        x = 2.0 * k / self.kmx - 1.0
        ex = self.emb(x)

        outs = []
        for b in self.b_lst:
            rj = b(ex)
            fj = self.f0 + (self.f1 - self.f0) * torch.sigmoid(rj)
            outs.append(fj)

        # NO SORT
        return torch.cat(outs, dim=1)

def o_loss(fq, m_gap=10.0):
    df = fq[:, 1:] - fq[:, :-1] 
    v = torch.relu(m_gap - df)
    return torch.mean(v**2)

def fb_p_loss(phys, fq, kx, ky):
    # agg phys loss
    ls = []
    for j in range(fq.shape[1]):
        fj = fq[:, j:j+1]
        lj = phys.physics_loss(fj, kx, ky)
        ls.append(lj)

    ls = torch.stack(ls)
    return torch.mean(ls) + 0.25 * torch.max(ls)

km = np.pi / a

phys = TorchPhysicsResidual(
    a=a, b=b, eta=eta, rho=rho, D=D, h=h, ks=k_s, m_u=m_u, 
    N_T=N_T, sqrt_damping=sq_dmp, denom_damping=dn_dmp, 
    resonance_exclusion_hz=res_exc
).to(dv)

mdl = P5BPinn(
    kmx=km, f0=f0, f1=f1, brs=brs, hdn=256
).to(dv)

opt = torch.optim.AdamW(mdl.parameters(), lr=lr, weight_decay=w_d)
sch = torch.optim.lr_scheduler.StepLR(opt, step_size=300, gamma=0.5)

bst_pth = "bst_p5b_pinn.pt"

print("mdl init ok")
```

```python
def trn_ep_5b(
    mdl, opt, ldr, phys, ep, 
    lam_p=1.0, lam_o=1e-3, m_gap=15.0, 
    g_clip=0.5, g_warn=1000.0, g_stop=1e6
):
    mdl.train()
    
    t_L = 0.0
    p_L = 0.0
    o_L = 0.0
    n_st = 0

    for b_idx, kb in enumerate(ldr):
        kb = kb.to(dv, non_blocking=True)
        kx, ky = kb[:, 0:1], kb[:, 1:2]

        opt.zero_grad(set_to_none=True)
        fq = mdl(kb)

        l_p = fb_p_loss(phys, fq, kx, ky)
        l_o = o_loss(fq, m_gap)
        l_t = lam_p * l_p + lam_o * l_o

        if not torch.isfinite(l_t):
            dump_grads(mdl)
            dbg_fail(mdl, ep, b_idx, kb, fq, phys, l_p, l_o, l_t)
            raise RuntimeError(f"NaN loss @ ep:{ep} b:{b_idx}")

        l_t.backward()
        t_nrm = torch.nn.utils.clip_grad_norm_(mdl.parameters(), g_clip)

        if not torch.isfinite(t_nrm):
            dump_grads(mdl)
            dbg_fail(mdl, ep, b_idx, kb, fq, phys, l_p, l_o, l_t)
            raise RuntimeError(f"NaN grad @ ep:{ep} b:{b_idx}")

        if t_nrm.item() > g_warn:
            print(f"WARN | ep:{ep} b:{b_idx} | nrm:{t_nrm.item():.2e} clip:{g_clip:.2e}")

        if t_nrm.item() > g_stop:
            dump_grads(mdl)
            dbg_fail(mdl, ep, b_idx, kb, fq, phys, l_p, l_o, l_t)
            raise RuntimeError(f"BOOM grad: {t_nrm.item():.2e}")

        opt.step()

        t_L += l_t.item()
        p_L += l_p.item()
        o_L += l_o.item()
        n_st += 1

    return t_L / n_st, p_L / n_st, o_L / n_st


@torch.no_grad()
def val_ep_5b(mdl, ldr, phys, lam_p=1.0, lam_o=1e-3, m_gap=15.0):
    mdl.eval()
    
    t_L = 0.0
    p_L = 0.0
    o_L = 0.0
    n_st = 0

    for kb in ldr:
        kb = kb.to(dv, non_blocking=True)
        kx, ky = kb[:, 0:1], kb[:, 1:2]

        fq = mdl(kb)

        l_p = fb_p_loss(phys, fq, kx, ky)
        l_o = o_loss(fq, m_gap)
        l_t = lam_p * l_p + lam_o * l_o

        if not torch.isfinite(l_t):
            raise RuntimeError("NaN in val")

        t_L += l_t.item()
        p_L += l_p.item()
        o_L += l_o.item()
        n_st += 1

    return t_L / n_st, p_L / n_st, o_L / n_st
```

```python
hst = {
    "train_total": [], "train_phys": [], "train_order": [],
    "val_total": [], "val_phys": [], "val_order": []
}

bst_vl = float("inf")

print("start trn p5b")
print("L = Lp + Lo")

for ep in range(epx):
    t_t, t_p, t_o = trn_ep_5b(
        mdl=mdl, opt=opt, ldr=dl_trn, phys=phys, ep=ep,
        lam_p=lam_p, lam_o=lam_o, m_gap=m_gap,
        g_clip=g_clip, g_warn=g_warn, g_stop=g_stop
    )

    v_t, v_p, v_o = val_ep_5b(
        mdl=mdl, ldr=dl_val, phys=phys, 
        lam_p=lam_p, lam_o=lam_o, m_gap=m_gap
    )

    sch.step()

    hst["train_total"].append(t_t)
    hst["train_phys"].append(t_p)
    hst["train_order"].append(t_o)
    hst["val_total"].append(v_t)
    hst["val_phys"].append(v_p)
    hst["val_order"].append(v_o)

    if v_t < bst_vl:
        bst_vl = v_t
        torch.save({
            "epoch": ep,
            "model_state_dict": mdl.state_dict(),
            "val_total": float(v_t),
            "val_phys": float(v_p),
            "val_order": float(v_o),
            "n_branches": brs,
            "min_gap_hz": m_gap,
            "lambda_order": lam_o,
        }, bst_pth)

    if ep % 25 == 0:
        print(f"ep:{ep:04d} | trn[{t_t:.2e} {t_p:.2e} {t_o:.2e}] | val[{v_t:.2e} {v_p:.2e} {v_o:.2e}] | bst:{bst_vl:.2e}")

print("trn done")
print(f"bst val: {bst_vl:.2e}")
```

```python
@torch.no_grad()
def pred_rt_5b(mdl, phys, kx_rt, ky_rt, r_tol=0.25, res_gap=15.0):
    mdl.eval()

    kr = torch.tensor(
        np.column_stack([kx_rt, ky_rt]), 
        dtype=torch.float32, 
        device=dv
    )
    
    kx, ky = kr[:, 0:1], kr[:, 1:2]
    fq = mdl(kr)
    xs = np.arange(len(kx_rt))

    ax, ay, ab, ar = [], [], [], []
    fq_np = fq.detach().cpu().numpy()

    print("br preds pre-flt:")
    for j in range(fq.shape[1]):
        fj = fq[:, j:j+1]
        r_abs = torch.abs(phys.residual_norm(fj, kx, ky))

        fj_n = fj.detach().cpu().numpy().reshape(-1)
        r_n = r_abs.detach().cpu().numpy().reshape(-1)

        fin = np.isfinite(fj_n) & np.isfinite(r_n)
        g_res = r_n < r_tol
        aw_res = np.abs(fj_n - phys.f_r) > res_gap

        msk = fin & g_res & aw_res

        ax.extend(xs[msk])
        ay.extend(fj_n[msk])
        ab.extend([j] * np.sum(msk))
        ar.extend(r_n[msk])

        print(
            f"br{j} | f:[{np.nanmin(fj_n):.1f}, {np.nanmax(fj_n):.1f}] avg:{np.nanmean(fj_n):.1f} | "
            f"r_med:{np.nanmedian(r_n):.2e} r_min:{np.nanmin(r_n):.2e} r_max:{np.nanmax(r_n):.2e} | "
            f"kpt:{np.sum(msk)}/{len(msk)}"
        )

    return (
        np.array(ax),
        np.array(ay),
        np.array(ab),
        np.array(ar),
        fq_np
    )

pn_x, pn_y, pn_b, pn_r, fq_rt_all = pred_rt_5b(
    mdl=mdl,
    phys=phys,
    kx_rt=rt_x,
    ky_rt=rt_y,
    r_tol=flt_tol,
    res_gap=flt_gap
)

print(f"tot kpt: {len(pn_x)}")
```

```python
def get_rts(fn, grd, arg, r_tol=1.0):
    v = np.array([fn(f, *arg) for f in grd])
    rts = []

    for j in range(len(grd) - 1):
        v0, v1 = v[j], v[j + 1]

        if not np.isfinite(v0) or not np.isfinite(v1):
            continue

        if np.sign(v0) != np.sign(v1):
            try:
                r = brentq(
                    fn, grd[j], grd[j + 1], 
                    args=arg, xtol=1e-6, maxiter=200
                )

                if abs(fn(r, *arg)) < r_tol:
                    if abs(r - f_r) > flt_gap:
                        rts.append(r)

            except Exception:
                pass

    rts.sort()
    u_rts = []

    for r in rts:
        if not u_rts or abs(r - u_rts[-1]) > 1e-3:
            u_rts.append(r)

    return u_rts


def calc_num_pts(clc, rx, ry, f_g):
    px, py = [], []

    for i, (kx, ky) in enumerate(zip(rx, ry)):
        if i % 30 == 0:
            print(f"k-pt {i}/{len(rx)}")

        rt = get_rts(clc.F_meta, f_g, (kx, ky), r_tol=1.0)

        for f in rt:
            px.append(i)
            py.append(f)

    return np.array(px), np.array(py)


clc = FCalculator(
    a=a, b=b, eta=eta, rho=rho, D=D, h=h, 
    ks=k_s, m_u=m_u, N_T=N_T
)

fg = np.linspace(f0, f1, 3000)

nm_x, nm_y = calc_num_pts(
    clc=clc,
    rx=rt_x,
    ry=rt_y,
    f_g=fg
)

print(f"num rts ok: {len(nm_x)}")
```
